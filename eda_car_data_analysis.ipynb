```python
url = 'https://raw.githubusercontent.com/wmaldonadodsa/analise-de-dados-e-boas-praticas/main/CAR%20DETAILS%20FROM%20CAR%20DEKHO.csv'

# Informa o cabeçalho das colunas
colunas = ['name', 'year', 'selling_price', 'km_driven', 'fuel', 'seller_type', 'transmission', 'owner']

# Lê o arquivo utilizando as colunas informadas
df = pd.read_csv(url, names=colunas, skiprows=1, delimiter=',')

df.head()
```
